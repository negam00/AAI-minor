```python
import pandas as pd

train_csv = pd.read_csv("train.csv", sep=",")

fields = [
          "Flight Distance", "Departure Delay in Minutes", 
          "Arrival Delay in Minutes", "Age"
          ]

def calculate_average(data, key):
  return data[key].mean()

def calculate_modus(data, key):
  return data.mode()[key][0]

def calculate_median(data, key):
  return data[key].median()

def generate_hist(data, key):
  return data.hist(column=key)
  
print("Gemiddelde:")
for field in fields:
  print(
      f"{field}: {calculate_average(train_csv, field)}"
  )

print("\nModus:")
for field in fields:
  print(
      f"{field}: {calculate_modus(train_csv, field)}"
  )

print("\nMediaan:")
for field in fields:
  print(
      f"{field}: {calculate_median(train_csv, field)}"
  )

print("\nVerdeling:")
for field in fields:
  generate_hist(train_csv, field)
```
