 Laat de foto op het scherm zien. Hoeveel gezichten zie jij?

```python
#Import image
img = plt.imread('sample_data/dag4_crowd.jpg')

cv.waitKey(0)
cv.destroyAllWindows()

#Show the image with matplotlib
plt.figure(figsize=(12,10))
plt.imshow(img)
plt.show();
plt.axis('off');
```

## Initieer detector

Download de .xml file met de classifier: haarcascade_frontalface_default.xml

(Het kan zijn dat deze file al meegekomen is met opencv)

Initieer de Cascade Classiefier.

```python
face_cascade = cv.CascadeClassifier(
    'sample_data/haarcascade_frontalface_default.xml')

print(f"Empty yes/no: {str(face_cascade.empty())}")
```

## Zet een kleurenplaatje om in een grijswaarde plaatje.

Als je foto al zrwrt-wit is dan kun je deze stap overslaan.

```python
grey_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.figure(figsize=(15,10))
plt.imshow(grey_img, cmap=('gray'))
plt.show();
plt.axis('off');
```

## Detecteer gezichten

Bepaal met de classiefier de gezichten in het plaatje.

Hoeveel heb je er gevonden?

```python
classified_faces = face_cascade.detectMultiScale(
    grey_img,
    scaleFactor=1.2,
    minNeighbors=5,
    minSize=(10,10),
    flags=cv.CASCADE_SCALE_IMAGE
)

print(f"Total amount of faces is {classified_faces.shape[0]}")
```

## Toon rechthoeken rond gezichten

Toon rechthoeken om de gevonden gezichten.

```python
for (x, y, w, h) in classified_faces:
  cv.rectangle(grey_img, (x,y), (x+w, y+h), (255, 0,0), 3)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(grey_img, cmap=('gray'))
plt.show();
plt.axis('off');
```
